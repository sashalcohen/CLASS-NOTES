# Data Aggregation/Summary Analysis

```Data aggregation``` is the process by which we take data across rows or down columns and express it in a summary statistical form.

The goal is to provide a high-level snapshot of outcomes or trends for a broad population, such as an entire country (or state or city), or for an industry.

A ```summary analysis```, calculates measures of central tendency (mean, median, mode) or variation (range, standard deviation).

Imagine this mock spreadsheet has thousands of rows of employment data:

But ```aggregation``` can mask important differences between subgroups and does not capture the full diversity of experiences within a population.





The opposite of ```aggregation``` is ```disaggregation``` which involves breaking down data into smaller, more specific subgroups based on characteristics like income levels, gender, race, or other relevant demographics.

```Disaggregated``` data is critical in surfacing inequities and disparities. **More on this next week**
